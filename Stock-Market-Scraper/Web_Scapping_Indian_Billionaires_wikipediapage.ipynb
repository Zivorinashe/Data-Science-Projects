# Web Scrapping

### Step 1: Fetch the web page and convert the HTML page into text with the help of the Python request library

```python
#import the python request
import pandas as pd
import requests
#the url for the website
Link = "https://en.m.wikipedia.org/wiki/Forbes_India%27s_richest_Indian_by_year"
#convert the web page to text
Link_text = requests.get(Link).text
print(Link_text)
```

### Step 2: To fetch useful data, convert Link_text into a BeautifulSoup. 

```python
#Import BeautifulSoup library from bs4 to pull data out of HTML and XML files
from bs4 import BeautifulSoup
#convert Link_text into a BeautifulSoup
soup = BeautifulSoup(Link_text, 'lxml')
print(soup)
```

### Step 3: the prettify() function make the indentation proper



```python
print(soup.prettify())
```

### Step 4:  use soup.title to fetch the web page title

```python
print(soup.title)
```

### Step 5: Diplaying webpage title (string) without tags

```python
print(soup.title.string)
```

### Step 6: explore <a></a> tags in the soup object



```python
#Displaying first <a></a> tag
soup.a
```

### Step 7: explore all <a></a> tags in the soup object


```python
# Displaying all <a></a> tags
soup.find_all('a')
```

### Step 8: Displaying all table tags

```python
#show table tags
all_table = soup.find_all('table')
print(all_table)
```

### Step 9:  Go to the webpage,find out the table class name.

### step 10: Fetch all table tags with the class name ‘wikitable sortable’

```python
our_table = soup.find('table', class_= 'wikitable sortable')
print(our_table)
```

### Step 11: Find <a></a> tags in the table

```python
table_links = our_table.find_all('a')
print(table_links)
```

### Step 12: Use the get() function to put the title on a list, iterate over table_links and append the title 

```python
billionaires = []
for links in table_links:
    my_title = links.get('title')
    if (my_title != None):
        billionaires.append(links.get('title'))
print(billionaires)
```

### Step 13: The data is in the form of a list, we will be using Python Pandas library to save the data in an Excel file

```python
#Convert the list into a dataframe 
df = pd.DataFrame(billionaires)
print(df)
```

### Step 14: Write the data into an Excel sheet

```python
#save the data into an excel file
writer = pd.ExcelWriter('indian_billionaires.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='List')
writer.save()
```
