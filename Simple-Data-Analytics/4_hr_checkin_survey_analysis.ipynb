```python
# Import important packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
import math
```

```python
data = pd.read_excel("HR Check-In Survey.xlsx")
data.head()
```

```python
# Set the font sizes
plt.rcParams["font.size"] = 12  # Title font size
plt.rcParams["axes.labelsize"] = 12  # Axis labels font size
plt.rcParams["xtick.labelsize"] = 12  # X axis ticks labels font size
plt.rcParams["ytick.labelsize"] = 12  # Y axis ticks labels font size
```

```python
import matplotlib.pyplot as plt

def horizontal_barplot(column: pd.Series, graph_title: str, x_label: str, y_label: str) -> None:
    """
    Definition of function:
    column: dataset['column name']
    graph_title: 'Title of graph'
    x_label: 'Name of x label you want'
    y_label: 'Name of y label you want'
    """
    # Count number of occurrences of each response
    value_counts = column.value_counts()

    # Create a figure and axes object
    fig, ax = plt.subplots()

    # Plot the bar chart
    value_counts.sort_index().plot(kind="barh", color=['#0091ff','#17d7a0','#4352e2', '#ff8181','#FDB353', '#4A5AE2'], x="count", ax=ax)
    

    # Set the title and labels
    plt.title(graph_title, fontsize=13)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.xlim(0, 30, 5)
    

    # Label the bars
    for i, v in enumerate(value_counts.sort_index().values):
        ax.text(v + 0.05, i, f"{v} ({round(100 * v / column.shape[0])}%)")

    # Show the plot
    plt.show()

```

```python
horizontal_barplot(data['Do you feel you have the resources needed to fulfill your duties effectively?'],
                   'Adequacy of Resources for Effective Duty Fulfillment', 'Count', '')
```
