# Data Visualization for Data Analysis

## 1. Introduction To Plotting In Python

Data plotting/ visualization is a very useful asset in presenting key findings in your data. Data visaulization has a few key objectives:
* To visualize relationships between data points
* To show variation or spread in data
* To summarize large, complex or overwhelming sets of numbers

If your plotting does not meet the above objectives, it is likely that a plot was not needed and that the data/ numbers could be described directly without a plot. However, after one decides to plot, there are a some recommended guidelines to ensure consistent and greta looking plots:
* Set your fonts as variables, both font type and font size: title, x-label. y-label etc. to reduce variation between plots
* Set your color scheme as an array to reduce the intrduction of random colors by your plotting package
    * Color scheme must be in line with the organizational requirements/ guidelines
* Avoid long titles, keep them concise and straight to the point
    * Always center your plot titles
* Make sure to order your x-label and y-labels in ascending order (1, 2, 3 ... or Low, Medium, High)
* Present all information, even zero counts, so the audiance can have a full view of the information
* Ensure your visualizations are easy to understand. If you are having a difficult time making sense of your plot, it's  most likely the audience will have a hard time understanding it
* Consider using function to automate similar plots

### 1.2.1 Binary Variables
Binary variables have 2 possible outcomes, Yes/ No, 1/ 0, Tall/ Short etc. They are the simplest to plot. Consider the distribution of gender in the Heart Disease dataset.
* Using Matplotlib independently

```python
gender = heart_disease_df.male.value_counts().index
gender_totals = heart_disease_df.male.value_counts().values

# creating the bar plot
fig = plt.figure(figsize = (8, 5))
plt.bar(gender, gender_totals, color = colors, width = 0.4)
 
plt.xlabel("Genders", fontsize = axis_font)
plt.ylabel("Total", fontsize = axis_font)
plt.title("Distribution of Gender", fontsize = title_font)
plt.xticks(np.arange(min(gender), max(gender)+1, 1), fontsize = ticks_font)
plt.yticks(fontsize = ticks_font)
plt.show()
```

* Using Pandas Matplotlib integration:

```python
counts = heart_disease_df.male.value_counts()

fig = plt.figure(figsize = (8, 5))
counts.plot(kind = 'bar', x = counts.index, y = counts.values, color = colors)
plt.xlabel("Genders", fontsize = axis_font)
plt.ylabel("Total", fontsize = axis_font)
plt.title("Distribution of Gender", fontsize = title_font)
plt.xticks(np.arange(min(gender), max(gender)+1, 1), rotation = 0, fontsize = ticks_font)
plt.yticks(fontsize = ticks_font)
plt.show()
```

As you can see, both will do the same work. We will focus on using the Pandas intergration moving forward, but feel free to use whichever feels more comfortable for you.

### 1.2.2 Categorical Variables
Categorical variables are part of grouped data, but have more than 2 groups. Categorical data can be:
* Nominal: type of data used to name variables e.g red, blue, green
* Ordinal: data which has a scale or order to it e.g unimportant, neutral, important

Generally, when you plot nomial data, there is no need to order it. It can be a pie chart to show the distribution of color of cars in a garage or shoes in a store. However, with ordinal data, your plots need to be ordered to show the scale/ order of data e.g. distribution of categories of salaries per month, goes from $0-499, $500 - 999, $1000+.

Start by plotting the bar distribution of education (ordered) in the Heart Disease dataset:

```python
education = heart_disease_df.education.value_counts().index

# Get the counts for each education level
education_totals = heart_disease_df.education.value_counts().values

# Sort the education data by the total count
 # Get the indices of the sorted education totals
sorted_indices = education_totals.argsort() 
 # Sort the education array using the sorted indices
education = education[sorted_indices] 

# Create a bar plot
fig = plt.figure(figsize=(8, 5))
plt.bar(education, education_totals[sorted_indices], color = colors, width=0.4)

# Add labels and title
plt.xlabel("Education Level", fontsize=axis_font)
plt.ylabel("Total", fontsize=axis_font)
plt.title("Distribution of Education Levels", fontsize=title_font)
plt.xticks(education, fontsize=ticks_font)

# Set the y-axis ticks and labels
plt.yticks(fontsize=ticks_font)

# Show the plot
plt.show()
```

Now we can plot a pie chart of the same data to also show distribution:

```python
#Education Status: Pie Chart
counts = heart_disease_df.education.value_counts()

fig = plt.figure(figsize = (8, 5))
counts.plot(kind = 'pie',
    # show percentage with two decimal points
    autopct='%1.2f%%',
    # increase the size of all text elements
    textprops={'fontsize': ticks_font},
    colors = colors)

plt.legend();
```

### 1.2.3 Continuous Variables

Continuous data is data that can take any value. Height, weight, temperature and length are all examples of continuous data. For continuous data, we generally make use of histograms to assess distribution of the variable on its own, or scatter/ line plots to see how it varies with another variable.

Lets start of by making the histogram of age:

```python
fig = plt.figure(figsize = (8, 5))
heart_disease_df.age.plot(kind = 'hist', color = colors)
plt.xlabel("Age", fontsize = axis_font)
plt.ylabel("Frequency", fontsize = axis_font)
plt.title("Distribution of Age", fontsize = title_font)
plt.xticks(fontsize = ticks_font)
plt.yticks(fontsize = ticks_font);
```

How would you plot 2 histograms of age, overlapping on the same graph, for male and female participants in the Heart Disease dataset?

```python
#answer here
# overlapping histogram:

# Filter data for male and female participants
df_male = heart_disease_df[heart_disease_df["male"] == 1]
df_female = heart_disease_df[heart_disease_df["male"] == 0]

# Create a histogram for male participants
fig, ax1 = plt.subplots(figsize = (8, 5))
ax1.hist(df_male["age"], color="#ff8181", alpha=0.7, label="Male")
plt.legend(loc='upper left')

# Create a histogram for female participants on the same graph
ax2 = ax1.twinx()  # Create a second y-axis
ax2.hist(df_female["age"], color="#17d7a0", alpha=0.7, label="Female")

# Set labels and title
ax1.set_xlabel("Age", fontsize=axis_font)
ax1.set_ylabel("Frequency (Male)", color= "#ff8181")
ax2.set_ylabel("Frequency (Female)", color="#17d7a0")
plt.title("Age Distribution for Male and Female Participants", fontsize=title_font)
plt.legend(loc='upper right')
plt.show()
```

```python
#plot overlapping histograms for male and female participants #method (2)
def plot_overlapping_histograms(data, variable, title):
    plt.figure(figsize=(10, 6))
    
   # Filter data for male and female participants
    male_data = data[data['male'] == 1][variable]
    female_data = data[data['male'] == 0][variable]
    
    # Plot histograms for male and female participants
    plt.hist(female_data, bins=20, alpha=0.5, label='Female',color= "#17d7a0")
    plt.hist(male_data, bins=20, alpha=0.5, label='Male', color="#ff8181")
    
    # Add labels and title
    plt.xlabel('Age', fontsize=axis_font)
    plt.ylabel('Frequency', fontsize=axis_font)
    plt.title(title, fontsize=title_font)
    plt.xticks(fontsize = ticks_font)
    plt.yticks(fontsize = ticks_font)
    
    # Add legend or a key
    plt.legend(loc='upper right')
    
    # Show the plot
    plt.show()
# Plot overlapping histograms
plot_overlapping_histograms(heart_disease_df, 'age', 'Age Distribution for Male and Female Participants')
```

Now we would like to understand some additional metrics for age in general. Plot the boxplot for age? What does it tell you?

```python
#Answer here

heart_disease_df = pd.DataFrame(heart_disease_df)

# Create a boxplot for age
plt.figure(figsize=(8, 6))
plt.boxplot(heart_disease_df["age"])
plt.title("Boxplot of Age in Heart Disease Dataset",fontsize = title_font)
plt.xlabel("Age",fontsize = axis_font)
plt.ylabel("Value",fontsize = axis_font)
plt.xticks(fontsize = ticks_font)
plt.yticks(fontsize = ticks_font)
plt.show()


# What does it tell: boxplot of age in the heart disease dataset:

# 1.Median Age :
# The line inside the box represents the median age. It gives an idea of the central tendency of the age distribution.
#If the line is closer to the bottom or top of the box,
#it indicates whether the majority of ages are lower or higher, respectively.

# 2.Interquartile Range (IQR):
# The box represents the IQR, which contains the middle 50% of the ages.
#A larger box suggests a wider spread of ages within the dataset.

# 3.Whiskers: 
# The whiskers extend from the box to show the range of ages within a certain limit. 
#Points beyond the whiskers are potential outliers.
#If there are outliers, they could be individuals with ages significantly higher or lower than the majority.

# 4.Outliers:
# Individual points beyond the whiskers are considered outliers. 
#These are data points that fall outside the expected range and may warrant further investigation. 
#Outliers could be due to data entry errors, anomalies, or genuinely unique cases.

# By examining these features, you can gain insights into the central tendency, spread, and
#potential extreme values in the distribution of ages in the heart disease dataset. 
#If the box is skewed towards one end, it suggests a concentration of ages in that direction.
#Outliers may indicate unusual cases that could be interesting for further analysis or might need validation.
```

The last segment of this trainin will combine the last training on functions and data visualization. Below, make three functions, and plot any 1 of the data points plotted above. Create a function to plot:
* A bar chart (binary or categorical variable)
* A histogram chart (continuous variable)

The parameters of your function should have the following:
* data/ df
* the column (s) to be plotted
* an x label parameter (to name the x axis)
* a y label parameter (to name the x axis)
* a title parameter
* any other key information you feel necessary

```python
#Answer here:
# Bar Chart
# Function to plot a bar graph
def plot_bar_graph(data, category_column, x_label, y_label, title,
                   colors = ['#0091ff', '#17d7a0', '#4352e2', '#fdb353', '#ff8181']):
    plt.figure(figsize=(10, 6))
    data[category_column].value_counts().plot(kind='bar', color=colors)
    plt.xlabel(x_label,fontsize=axis_font)
    plt.ylabel(y_label,fontsize=axis_font)
    plt.xticks(rotation=0, fontsize=ticks_font)
    plt.yticks(fontsize=ticks_font)
    plt.title(title,fontsize=title_font)
    plt.show()

# Bar Graph
plot_bar_graph(heart_disease_df, 'education', 'education', 'Count', 'Education')
```

```python
#Answer here:
# Histogram Chart
# Function to plot a histogram
def plot_histogram(data, column, x_label, y_label, title, bins=10, title_font = 20,
    axis_font = 15 ,ticks_font = 15, bar_label_font = 15, color='#0091ff'):
    plt.figure(figsize=(10, 6))
    plt.hist(data[column], bins=bins, color=color)
    plt.xlabel(x_label,fontsize = axis_font)
    plt.ylabel(y_label,fontsize = axis_font)
    plt.xticks(fontsize = ticks_font)
    plt.yticks(fontsize = ticks_font)
    plt.title(title,fontsize = title_font)
    plt.show()

# Histogram
plot_histogram(heart_disease_df, 'age', 'age', 'Count', 'Gender')
```

```python
#Answer here:
# Scatter Plot
# Function to plot a scatter plot
def plot_scatter_plot(data, x_column, y_column, x_label, y_label, title, title_font = 20,
    axis_font = 15 ,ticks_font = 15, bar_label_font = 15,color='#4352e2'):
    plt.figure(figsize=(10, 6))
    plt.scatter(data[x_column], data[y_column], color=color)
    plt.xlabel(x_label,  fontsize=axis_font)
    plt.ylabel(y_label, fontsize=axis_font)
    plt.xticks(fontsize = ticks_font)
    plt.yticks(fontsize = ticks_font)
    plt.title(title, fontsize=title_font)
    plt.show()
    
# Scatter Plot
plot_scatter_plot(heart_disease_df, 'age', 'BMI', 'age', 'BMI', 'BMI by age')
```

```python
#Answer here:
# Pie Chart
# Function to plot a pie chart
def plot_pie_chart(data, category_column, title):
    plt.figure(figsize=(8, 8))
    title_font = 20
    axis_font = 15
    ticks_font = 15
    bar_label_font = 15
    data[category_column].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90,
        textprops={'fontsize': ticks_font},colors = ['#0091ff', '#17d7a0', '#4352e2', '#fdb353', '#ff8181'])
    plt.title(title, fontsize=title_font)
    plt.axis('equal') 
    plt.show()
    
# Pie Chart
plot_pie_chart(heart_disease_df, 'education', 'Education Distribution' )
```
