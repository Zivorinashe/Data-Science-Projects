```python
import matplotlib.pyplot as plt

def horizontal_barplot(column: pd.Series, graph_title: str, x_label: str, y_label: str) -> None:
    """
    Definition of function:
    column: dataset['column name']
    graph_title: 'Title of graph'
    x_label: 'Name of x label you want'
    y_label: 'Name of y label you want'
    """
    # Count number of occurrences of each response
    value_counts = column.value_counts()

    # Create a figure and axes object
    fig, ax = plt.subplots()

    # Plot the bar chart
    value_counts.sort_index().plot(kind="barh", color=['#0091ff','#17d7a0','#4352e2', '#ff8181','#FDB353', '#4A5AE2'], x="count", ax=ax)
    

    # Set the title and labels
    plt.title(graph_title, fontsize=13)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.xlim(0, 30, 2)
    

    # Label the bars
    for i, v in enumerate(value_counts.sort_index().values):
        ax.text(v + 0.05, i, f"{v} ({round(100 * v / column.shape[0])}%)")

    # Show the plot
    plt.show()

```

```python
horizontal_barplot(data['Have you received any training in teaching game coding to children before?'],
                   'Prior Training in Game Coding', 'Count', '')
```

```python
# Create the data
importance_levels = {
    "Yes": 14,
    "No": 0
}

# Convert the dictionary keys to a list
importance_level_list = list(importance_levels.keys())

# Extract the values from the dictionary
importance_level_values = list(importance_levels.values())

# Create the horizontal bar chart
fig, ax = plt.subplots()
ax.barh(importance_level_list, importance_level_values, color=['#0091ff', '#17d7a0', '#4352e2'])

# Add values and percentages to the left of each bar
for i, bar in enumerate(ax.patches):
    value = importance_level_values[i]
    percentage = round(value / sum(importance_level_values) * 100)
    plt.text(bar.get_x() + bar.get_width() + 0.09, bar.get_y() + bar.get_height() / 2, f'{value} ({percentage}%)', ha='left', va='center')

# Set the title and labels for the bar chart
plt.title('Understanding Key Features', fontsize=16)
plt.ylabel('')
plt.xlabel('Count')
plt.xlim(0,27,5)

# Display the bar chart
plt.show()
```

```python

# Create a new dictionary to store the number of respondents who selected each checkbox answer
checkbox_answer_counts = {}

# Iterate over the data and count the number of times each checkbox answer was selected
for respondent in data.index:
    for checkbox_answer in data['What are the best ways to teach game coding to students?   (Please select all that apply)'][respondent].split(','):
        # Skip empty checkbox answers
        if checkbox_answer.lstrip() == '':
            continue

        # Increment the count for the checkbox answer
        checkbox_answer_counts.setdefault(checkbox_answer.lstrip(), 0)
        checkbox_answer_counts[checkbox_answer.lstrip()] += 1

# Calculate the percentages for each checkbox answer
answer_percentages = {}
for answer in checkbox_answer_counts:
    answer_percentages[answer] = checkbox_answer_counts[answer] / len(data.index) * 100

# Create a bar chart of the number of respondents who selected each checkbox answer
answers = list(checkbox_answer_counts.keys())
values = list(checkbox_answer_counts.values())

# 
percentages = [round(value/sum(values) * 100) for value in values]
percentages_as_strings = [f'{percentage}%' for percentage in percentages]

fig, ax = plt.subplots(figsize=(10,4))
plt.barh(answers, values, color=['#0091ff','#17d7a0','#4352e2', '#ff8181','#FDB353', '#4A5AE2'] )

# Code to show both percentages and numbers the graphs
for i, bar in enumerate(ax.patches):
    plt.text(bar.get_width() / 1, bar.get_y() + bar.get_height() / 2, f' {values[i]} ({percentages_as_strings[i]})')
    

# Set the title and labels for the bar chart
plt.title('Effective Approaches to Game Coding Education  ', fontsize=16)
plt.xlabel('Count')
plt.xlim(0,23, 5)
plt.ylabel('')

# Display the bar chart
plt.show()
```
